```python
#Loading the symbols dataset
symbols = pd.read_csv(r"C:\Users\POORNIMA\Downloads\Symbols.csv")
```

```python
#Converting the array values to a single list
tickers_list = []
for t in list(symbols.values):
    print(t)
    x=t[0]
    tickers_list.append(x)
```

```python
#printing/checking company name for each of the symbol/stock
for t in list(symbols.values):
    print(t)
    x=t[0]
    company_name = yf.Ticker(x).info['longName']
    print(company_name)
```

```python
#Downloading all the stocks data
data = yf.download(tickers=tickers_list,group_by='ticker')
```
